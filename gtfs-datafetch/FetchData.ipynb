# Finding and fetching train stations in central europe

## GTFS
The General Transit Feed Specification (GTFS) allows public transit companies to publish their data in an way that it is universally readable and usable by developers all over the world. Currently the GTFS standard is used by thousands of public transport providers, including companies like SBB, SNFC or OEBB.

## Usage
In order to build a system that enables us to provide customers with the best connection in terms of time, time to switch trains or amount of changes, we need to know all the stops and connections. This data is provided by using the GTFS specification.

## Data soruces
A lot of countries and companies in the public sector have a open data policy, this means that they provide their data free of cost to everyone who wannts to use it. Some have more restrictive licensing than others. For a proof of concept every datasoruce is valid to use.

The bigest struggle was to find all the data soruces and bring them together in a central point for us to use. There are projects like https://openmobilitydata.org that already did a very good job in collecting these datasoruce. But by far not every country or provider have their data on this platform. To find these datasources a regular google search and clicking through goverment and carrier websites was needed.

It was possible to find data all of the countries we were instrested in - for some there was one dowload page - some needed some more downloading efforts.
